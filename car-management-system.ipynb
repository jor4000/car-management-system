```python
#You can create, show, add damages and delete damages for an objet(a car)
class Car:
    def __init__(self, brand, color, year):
        self.brand = brand
        self.color = color
        self.year = year
        self.damage = []

    def __str__(self):
        return f"Brand: {self.brand} | Color: {self.color} | Year: {self.year} | Damages: {[i["damage"].title() for i in self.damage]}"

    def add_damage(self, damage, description=""):
        self.damage.append({"damage": damage.lower(), "description": description})
        print(f"New damage added: [Damage: {damage.lower()} | Description: {description or None}]")

    def del_damage(self):
        print(f"Damages:")
        for index, value in enumerate(self.damage, start=1):
            print(f"{index}. {value["damage"].title()}")

        index = input("Index of the damage to be eliminated: ")
        if index.isdigit():
            index = int(index) - 1

            if 0 <= index < len(self.damage):
                removed = self.damage.pop(index)
                print(f"Deleted damage: [Damage: {removed["damage"].title()} | Description: {removed["description"] or None}]")

            else:
                print("Error: Index out of range")

        else:
            print("Error: Enter a valid number")

#Test
car_1 = Car("Porsche", "Orange", "2024")
print(car_1)
car_1.add_damage("Gearbox failure", "20000$ for repair")
car_1.add_damage("Leak in the water pump")
print(car_1)
car_1.del_damage()
```
